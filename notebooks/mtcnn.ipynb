```python
import cv2

def list_cameras():
    index = 0
    available_cameras = []
    while True:
        cap = cv2.VideoCapture(index)
        if cap.read()[0]:  # Check if the camera is available
            available_cameras.append(index)
            cap.release()  # Release the device
        else:
            break  # No more cameras available
        index += 1
    return available_cameras

# Get the list of available cameras
cameras = list_cameras()
print("Available camera indices:", cameras)
```

```python
import os
import cv2
from mtcnn import MTCNN
from PIL import Image
import numpy as np

# Create a detector instance
detector = MTCNN()

# Directory containing images
image_folder = "image/"  # Replace with your folder path
output_folder = "output/"  # Directory to save images with drawn boxes

# Ensure output directory exists
os.makedirs(output_folder, exist_ok=True)

# Loop through all images in the folder
for filename in os.listdir(image_folder):
    # Construct the full file path
    file_path = os.path.join(image_folder, filename)

    # Check if the file is an image
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
        print(f"Processing {file_path}...")

        # Load the image using PIL and convert to numpy array
        image = Image.open(file_path)
        image_np = np.array(image)

        # Detect faces in the image
        results = detector.detect_faces(image_np)

        # Convert the image back to OpenCV format (BGR for visualization)
        image_cv = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)

        # Draw the bounding box and keypoints for each detected face
        for result in results:
            # Bounding box
            x, y, width, height = result['box']
            cv2.rectangle(image_cv, (x, y), (x + width, y + height), (0, 255, 0), 2)  # Green box

            # Keypoints
            keypoints = result['keypoints']
            for point in keypoints.values():
                cv2.circle(image_cv, point, 5, (0, 0, 255), -1)  # Red keypoints

        # Save the image with the drawn boxes
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, image_cv)
        print(f"Saved processed image to {output_path}")

print("Processing complete.")
```

```python
import os
import cv2
from mtcnn import MTCNN
from PIL import Image
import numpy as np
import math

# Create a detector instance
detector = MTCNN()

# Directory containing images
image_folder = "image/"  # Replace with your folder path
output_folder = "output/"  # Directory to save images with drawn boxes

# Ensure output directory exists
os.makedirs(output_folder, exist_ok=True)

# Known real-world distance between eyes (in cm)
REAL_EYE_DISTANCE = 6.3  # Average human interpupillary distance
FOCAL_LENGTH = 700  # Focal length of the camera (in pixels)

# Function to calculate distance from eye keypoints
def calculate_distance(left_eye, right_eye, focal_length, real_eye_distance):
    pixel_distance = math.sqrt((left_eye[0] - right_eye[0])**2 + (left_eye[1] - right_eye[1])**2)
    distance = (focal_length * real_eye_distance) / pixel_distance
    return distance

# Loop through all images in the folder
for filename in os.listdir(image_folder):
    file_path = os.path.join(image_folder, filename)

    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
        print(f"Processing {file_path}...")

        try:
            # Load the image using PIL and convert to numpy array
            image = Image.open(file_path)
            image_np = np.array(image)

            # Detect faces in the image
            results = detector.detect_faces(image_np)

            # Convert the image back to OpenCV format (BGR for visualization)
            image_cv = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)

            # Draw the bounding box and keypoints for each detected face
            for result in results:
                x, y, width, height = result['box']
                cv2.rectangle(image_cv, (x, y), (x + width, y + height), (0, 255, 0), 2)  # Green box

                keypoints = result['keypoints']
                for point in keypoints.values():
                    cv2.circle(image_cv, point, 5, (0, 0, 255), -1)  # Red keypoints

                # Calculate and display distance if both eyes are detected
                if 'left_eye' in keypoints and 'right_eye' in keypoints:
                    left_eye = keypoints['left_eye']
                    right_eye = keypoints['right_eye']
                    distance = calculate_distance(left_eye, right_eye, FOCAL_LENGTH, REAL_EYE_DISTANCE)
                    print(f"Distance to face: {distance:.2f} cm")
                    cv2.putText(
                        image_cv, f"Dist: {distance:.2f} cm", 
                        (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                        0.6, (255, 255, 255), 2
                    )

            # Save the image with the drawn boxes
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, image_cv)
            print(f"Saved processed image to {output_path}")

        except Exception as e:
            print(f"Error processing {file_path}: {e}")

print("Processing complete.")
```
